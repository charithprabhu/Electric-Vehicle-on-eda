```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your dataframe and 'Model Year' is a column in it
plt.figure(figsize=(10, 6))  # Corrected figsize format

# Plot a histogram with a KDE (Kernel Density Estimate)
sns.histplot(df['Model Year'], bins=20, kde=True, color='#0000FF', edgecolor='black')

# Title of the plot
plt.title('Distribution of Model Year', fontsize=16, fontweight='bold', color='#111')

# Labels for the x and y axes
plt.xlabel('Model Year', fontsize=12, fontweight='medium')
plt.ylabel('Frequency', fontsize=12, fontweight='medium')

# Adjust ticks on the x and y axes
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Adding a grid with transparency and linestyle
plt.grid(True, linestyle='-', alpha=0.6)

# Display the plot
plt.show()
```

Create a Choropleth using plotly.express to display the number of EV vehicles based on location.

```python
import plotly.express as px

# Group data by state and count the number of EVs
state_data = df.groupby('State')['VIN (1-10)'].count().reset_index()
state_data.columns = ['State', 'EV Count']  # Rename columns for clarity

# Create the choropleth map
fig = px.choropleth(
    state_data,
    locations='State',
    locationmode='USA-states',
    color='EV Count',
    color_continuous_scale='Purples',
    labels={'EV Count': 'Number of EV Vehicles'},
    title='Number of Electric Vehicles by State'
)

# Show the map
fig.show()
```

```python
import plotly.express as px

# Group the data by state and count the number of EVs
count_by_state = df.groupby("State").size().reset_index(name="count")

# Create the choropleth map
fig = px.choropleth(
    count_by_state,
    locations="State",
    color="count",
    locationmode="USA-states",
    scope="usa",
    labels={"count": "NO OF EV's"},
    title="Number of EVs by State"
)

# Update layout with custom settings
fig.update_layout(
    geo=dict(
        bgcolor='rgba(0,0,0,0)'  # Transparent background for the map
    ),
    title_x=0.5  # Center the title
)

# Show the map
fig.show()
```

```python
import plotly.express as px

# Group the data by 'Electric Vehicle Type' and count the occurrences
vehicle_type_count = df.groupby('Electric Vehicle Type')['VIN (1-10)'].count().reset_index()
vehicle_type_count.columns = ['Electric Vehicle Type', 'Count']  # Rename the columns for clarity

# Create the pie chart
fig = px.pie(
    vehicle_type_count,
    names='Electric Vehicle Type',
    values='Count',
    title='Distribution of Electric Vehicles by Type',
    labels={'Electric Vehicle Type': 'EV Type', 'Count': 'Number of Vehicles'},
    hole=0.3,  # This makes it a donut chart
    color_discrete_sequence=px.colors.sequential.Plasma  # Use the Plasma color sequence
)

# Update the trace to show percentage and label
fig.update_traces(textinfo='percent+label')

# Update the layout to set title font size and adjust margins
fig.update_layout(
    title_font_size=20,
    margin=dict(l=40, r=40, t=40, b=40)
)

# Show the pie chart
fig.show()
```

```python
import plotly.express as px
import warnings

# Group the data by 'Model Year' and 'Make' and count the occurrences
make_year_data = df.groupby(['Model Year', 'Make'])['VIN (1-10)'].count().reset_index()
make_year_data.columns = ['Model Year', 'Make', 'Count']  # Rename columns for clarity

# Create the animated bar chart
fig = px.bar(
    make_year_data,
    x='Make',
    y='Count',
    color='Make',
    animation_frame='Model Year',
    animation_group='Make',
    range_y=[0, make_year_data['Count'].max() + 50],
    title='Electric Vehicle Makes Over Time',
    labels={'Count': 'Number of Vehicles', 'Make': 'EV Make'},
    text='Count'
)

# Update layout for better presentation
fig.update_layout(
    title_font_size=20,
    xaxis_title_font_size=14,
    yaxis_title_font_size=14,
    legend_title_text='Electric Vehicle Make',
    margin=dict(l=40, r=40, t=40, b=40)  # Corrected margin settings
)

# Update traces to customize text position and format
fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')  # Adjusted formatting

# Show the animated bar chart
fig.show()
```
